## O(n) Algorithm Example

### Called Proportional 
Big O or O always refers to worst case scenario

```python

# Algorithm will execute at worst n times
# i.e. O(n)

# Also good to note, it will also run n times at best but thats only because we want to iterate through full list in this case
# So worst case = average case = best case or O(n) = Theta(n) = Omega(n)
def print_items(n):
    for i in range(n): 
        print(i)

print_items(4)
```

## Drop Constants: Big O

```python
# Below algorithm has two for loops (n + n)
# i.e. O(n+n) or O(2n) 
# We do not worry about constants in time complexity, we simplify
# This algorithm is still O(n)

def print_items(n):
    for i in range(n): 
        print(i)

    for j in range(n): 
        print(j)
print_items(4)
```

## O(n^2)

### Called Loop within a Loop 

```python
# Nested for loops (n*n)
# O(n^2)
# Less efficient than O(n) because many more operations
def print_items(n):
    for i in range(n): 
        for j in range(n): 
            print(i,j)
print_items(4)
```

## Drop Non Dominants

```python
# Below algorithm is n * n + n
# i.e. O(n^2+n)

# n^2 will grow much faster than n for large n's
# proportionally, that n will not mean much compared to n^2 
# therefore we drop n or non dominants
# The below algorithm has a time complexity of O(n^2) since we drop n

def print_items(n):
    for i in range(n): 
        for j in range(n): 
            print(i,j)

    for k in range(n):
        print(k)

print_items(4)
```

## O(1)

### Constant Time

```python
# In this type of algorithm, only 1 operation is done
# There is only one operation

def add_n(n):
    return n + n

add_n(3)
```

```python
# Even if there is two additions or multiple operations like in below
# The algorithm won't be O(3), it is still O (1) as O(1) represents constant time i.e. an algorithm that always has the same execution time

def add_subtract(n):
    return n + n - 2 -4

add_subtract(5)
```

## Different parameters in function

```python
# Since we are iterating over two numbers and not just "n"
# Our algorithm is O(a+b)
# It cannot be simplified further!!

def print_items(a,b): 
    for i in range(a): 
        print(i)

    for k in range(b):
        print(k)



# Similarly, the below algorithm is O(a*b)
# It cannot be simplified further!!
def print_items(a,b): 
    for i in range(a): 
        print(i)
        for k in range(b):
            print(k)
```

## List

```python
## Adding/remove to end of list is O(1) because no re-indexing

l = [1,2,3,4]

# O(1)
l.append(5)
# O(1)
l.pop()
l
```

```python
## Adding/removing from beginning is O(n) because re-indexing

l = [1,2,3,4]

# O(n)
l.insert(0, 11)

# O(n)
l.pop(0)

l
```

```python
## Adding/removing from anywhere in middle is O(n)
## May think it is O(1/2 * n) since we only reindex half or portion of list but drop constants!
## Therefore O(n)

l.insert(4, 11)
l
```
